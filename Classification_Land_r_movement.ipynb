## Pre-processing - smoothing filter to remove the electrical nosie

```python
def momentum(z):
    n_iter = len(z)
    beta = 0.9
    temp = np.zeros(len(z))
    
    for k in range(0,n_iter-1):
        temp[k+1] = beta*temp[k] +(1-beta) *z[k]
        
    return temp
```

### Bandpass Filter- having low pass frequency: 0.5Hz and higher frequency- 90Hz

```python
#creating filter variables
fs = 200.0 #200 hz sampling rate
lowcut = 0.5 #0.1hz is the lowest frequency we will pass
highcut = 70.0 #40hz is the highest frequency we will pass
```

```python
def butter_bandpass_filter(raw_data, fs, lowcut, highcut, order = 5):
    nyq = 0.5 *fs
    low = lowcut /nyq
    high = highcut /nyq
    sos = butter(order,[low,high], analog = False, btype = 'band', output = 'sos')
    filted_data = sosfiltfilt(sos, raw_data)
    return filted_data
```

# Time domain feature extraction

### Moving averages having 160 samples per second, considering window size of 2 secs ie 320 samples

```python
def moving_mean(sig,win_size):
    window_size = win_size

    i = 0
    moving_averages = []
    while i < len(sig) - window_size + 1:
        this_window = sig[i : i + window_size]

        window_average = sum(this_window) / window_size
        moving_averages.append(window_average)
        i += 1

    return(moving_averages)
```

### Variance having 160 samples per second, considering window size of 2 secs ie 320 samples

```python
def moving_variance(sig,win_size):
    window_size = win_size
    i = 0
    variance = []
    while i < len(sig) - window_size + 1:
        this_window = sig[i : i + window_size]

        window_variance = np.var(this_window)
        variance.append(window_variance)
        i += 1
    return(variance)
```

```python
import scipy
from scipy.stats import skew
def moving_skewness(sig, win_size):
    window_size = win_size
    i = 0
    skewness = []
    while i < len(sig) - window_size + 1:
            this_window = sig[i : i + window_size]

            window_skewness = scipy.stats.skew(this_window)
            skewness.append(window_skewness)
            i += 1
    return(skewness)    
```

```python
from scipy.stats import kurtosis
def moving_kurtosis(sig, win_size):
    window_size = win_size
    i = 0
    kur = []
    while i < len(sig) - window_size + 1:
            this_window = sig[i : i + window_size]

            window_kurtosis = scipy.stats.kurtosis(this_window)
            kur.append(window_kurtosis)
            i += 1
    return(kur)    
```

### Zero crossing
Zero crossings are a basic property of an audio signal that is often employed in audio classification. Zero crossings allow for a rough estimation of dominant frequency and the spectral centroid (SC) 

```python
import librosa
def zero_crossing(sig, win_size):
    window_size = win_size
    i = 0
    zero_crossing = []
    while i < len(sig) - window_size + 1:
            this_window = sig[i : i + window_size]

            window_zero_crossing = librosa.feature.zero_crossing_rate(this_window)
            zero_crossing.append(window_zero_crossing)
            i += 1
    return(zero_crossing)
```

```python
def moving_area_under_curve(sig,win_size):
    window_size = win_size

    i = 0
    area_under_curve = []
    while i < len(sig) - window_size + 1:
        this_window = sig[i : i + window_size]

        window_area = np.trapz(this_window)
        area_under_curve.append(window_area)
        i += 1

    return(area_under_curve)
```

### peak- to- peak signal

```python
def peak_to_peak_signal(sig,win_size):
    window_size = win_size

    i = 0
    peak_to_peak_signal = []
    while i < len(sig) - window_size + 1:
        this_window = sig[i : i + window_size]

        window_ppk = max(this_window) - min(this_window)
        peak_to_peak_signal.append(window_ppk)
        i += 1

    return(peak_to_peak_signal)
```

```python
from tensorflow.keras import regularizers
def get_regularised_model(input_shape, dropout_rate, weight_decay):
    model = tf.keras.Sequential([
        Dense(64,kernel_regularizer = regularizers.l2(weight_decay), 
              kernel_initializer='he_uniform', bias_initializer="ones",
              activation='relu', input_shape=(input_shape)),
        Dense(128,kernel_regularizer = regularizers.l2(weight_decay), activation='relu'),
        Dense(128,kernel_regularizer = regularizers.l2(weight_decay), activation='relu'),
        BatchNormalization(),
        Dense(128,kernel_regularizer = regularizers.l2(weight_decay), activation='relu'),
        Dense(128,kernel_regularizer = regularizers.l2(weight_decay), activation='relu'),
        Dropout(dropout_rate),
        Dense(64,kernel_regularizer = regularizers.l2(weight_decay), activation='relu'),
        Dense(64,kernel_regularizer = regularizers.l2(weight_decay), activation='relu'),
        Dropout(dropout_rate),
        Dense(64,kernel_regularizer = regularizers.l2(weight_decay), activation='relu'),
        Dense(64,kernel_regularizer = regularizers.l2(weight_decay), activation='relu'),
        Dense(4,activation='softmax')
    ])
    return(model)
```
